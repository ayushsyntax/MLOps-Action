# Git & GitHub for MLOps — A Clear and Practical Guide

Git is a **distributed version control system**.
GitHub is a **remote platform that hosts Git repositories**.

Together, they provide:

• Version tracking
• Collaboration
• Safe experimentation
• Reproducibility (critical for ML)

---

## 1. How Git Works (Mental Model)

Think of Git as a **four-stage pipeline** through which your code flows.

```
┌────────────┐
│ Workspace  │  ← You write or edit files here
└─────┬──────┘
      │ git add
      ▼
┌────────────┐
│ Staging    │  ← Files prepared for commit
│ Area       │
└─────┬──────┘
      │ git commit
      ▼
┌────────────┐
│ Local Git  │  ← Versioned history on your machine
│ Repository │
└─────┬──────┘
      │ git push
      ▼
┌────────────┐
│ GitHub     │  ← Remote backup & collaboration
└────────────┘
```

---

## 2. Core Git Workflow

### Step 1: Initialize Repository

```bash
git init
```

Creates a `.git/` directory that tracks everything.

---

### Step 2: Check File Status

```bash
git status
```

Shows:

* Untracked files
* Modified files
* Staged files

---

### Step 3: Stage Files

```bash
git add file.py
git add .
```

Moves files from **workspace → staging area**.

---

### Step 4: Commit Changes

```bash
git commit -m "Add data preprocessing logic"
```

Creates:

* Snapshot of code
* Unique SHA hash
* Permanent history record

---

### Step 5: Push to GitHub

```bash
git push origin main
```

Uploads commits to the remote repository.

---

## 3. Visualizing the Commit History

```
o---o---o---o   main
    \
     o---o      feature-branch
```

Each `o` is a commit (snapshot).

Git does **not** save diffs — it stores **complete states**.

---

## 4. Ignoring Files with `.gitignore`

Prevents unwanted files from being tracked.

Example:

```
.env
__pycache__/
data/
*.csv
*.log
```

Essential for:

* Secrets
* Datasets
* Temporary files
* Model artifacts

---

## 5. Branching (Core to MLOps)

Branching allows safe experimentation.

```
main
 |
 |---- feature-model-tuning
 |
 |---- feature-data-cleaning
```

### Commands

```bash
git branch feature-x
git checkout feature-x
```

Or shorter:

```bash
git checkout -b feature-x
```

---

## 6. Merging Branches

```
Before Merge:

main:     A ── B ── C
                 \
feature:          D ── E
```

After merge:

```
main:     A ── B ── C ── F
                     ↖── D ── E
```

Command:

```bash
git merge feature-x
```

---

## 7. Merge Conflicts (What They Look Like)

Occurs when two branches modify the same line.

```
<<<<<<< HEAD
model = RandomForest()
=======
model = XGBoost()
>>>>>>> feature-branch
```

Resolution steps:

1. Choose correct code
2. Remove markers
3. Save file
4. Commit

| ------- | --------------- |

| 1.1.0   | New feature     |

## 10. Git in MLOps Context

Git manages:

• Training code
• Feature engineering
• Pipelines
• Config files
• Experiment history

Git does **not** manage well:

• Large datasets
• Model binaries

(Use DVC, S3, MLflow, or artifact stores)

## 🧩 **Git's 4-Stage Architecture (ASCII Flow)**

```
┌─────────────┐     ┌──────────────┐     ┌─────────────────┐     ┌──────────────────┐
│             │     │              │     │                 │     │                  │
│   WORKSPACE │────▶│  STAGING     │────▶│  LOCAL          │────▶│  REMOTE          │
│   (Your     │     │  AREA        │     │  REPOSITORY     │     │  REPOSITORY      │
│   computer) │     │  (Prep area) │     │  (Your machine) │     │  (GitHub)        │
│             │     │              │     │                 │     │                  │
└─────────────┘     └──────────────┘     └─────────────────┘     └──────────────────┘
       │                   │                    │                    │
       ▼                   ▼                    ▼                    ▼
• Your project files  • git add           • git commit          • git push
• Untracked files     • "shopping cart"   • SHA-1 snapshots   • Cloud backup
• Active editing      • Select changes     • Local history     • Team access
```

### 💡 **Simple Analogy**
```
WORKSPACE    = Your messy desk with papers everywhere
STAGING AREA = Filing cabinet where you organize documents
LOCAL REPO   = Locked safe where you store finalized documents
GITHUB       = Shared cloud vault where your team can access files
```

## ⚙️ **Core Workflow: Step-by-Step Commands**

```
┌─────────────────────────────────────────────────────────────────────────────┐
│                     GIT WORKFLOW FOR DATA SCIENTISTS                        │
└─────────────────────────────────────────────────────────────────────────────┘

STEP 1: SETUP & CONFIGURATION
┌─────────────────────────────────────────────────────────────────────────────┐
│ git config --global user.name "Your Name"                                  │
│ git config --global user.email "your.email@company.com"                    │
│ git init  # Creates .git folder in your project directory                   │
└─────────────────────────────────────────────────────────────────────────────┘

STEP 2: TRACK FILES (The Cycle Begins)
┌─────────────────────────────────────────────────────────────────────────────┐
│ git status                  # Check current state (🔴 untracked, 🟢 staged)  │
│ git add data_preprocessing.py  # Stage specific file                       │
│ git add .                   # Stage ALL changes (be careful!)               │
│ git restore --staged config.py  # Unstage a file if needed                 │
└─────────────────────────────────────────────────────────────────────────────┘

STEP 3: SAVE CHANGES
┌─────────────────────────────────────────────────────────────────────────────┐
│ git commit -m "Add feature engineering for customer segmentation"         │
│ # Creates snapshot with unique SHA: a1b2c3d4e5f6...                         │
└─────────────────────────────────────────────────────────────────────────────┘

STEP 4: CLOUD SYNC
┌─────────────────────────────────────────────────────────────────────────────┐
│ git remote add origin https://github.com/yourusername/project.git          │
│ git push -u origin main  # Upload to GitHub, -u sets default branch        │
└─────────────────────────────────────────────────────────────────────────────┘
```

### **Branching Commands Cheat Sheet**
```
┌─────────────────────────────────────────────────────────────────────────────┐
│ # Create and switch to new branch                                           │
│ git branch feature-data-cleaning                                            │
│ git checkout feature-data-cleaning                                          │
│ # OR shortcut:                                                              │
│ git checkout -b feature-model-tuning                                        │
│                                                                             │
│ # Work on your branch (make commits)                                        │
│ git add .                                                                   │
│ git commit -m "Add data cleaning pipeline"                                 │
│                                                                             │
│ # Merge back to main                                                        │
│ git checkout main                                                           │
│ git pull origin main  # Get latest changes first                           │
│ git merge feature-data-cleaning                                             │
│                                                                             │
│ # If conflict occurs:                                                       │
│ # 1. Open conflicted files                                                  │
│ # 2. Look for <<<<<<< HEAD sections                                         │
│ # 3. Choose which code to keep                                              │
│ # 4. git add resolved_file.py                                               │
│ # 5. git commit -m "Resolve merge conflict"                                 │
└─────────────────────────────────────────────────────────────────────────────┘
```

## 🏷️ **Semantic Versioning with Tags**

```
┌─────────────────────────────────────────────────────────────────────────────┐
│                     VERSIONING TIMELINE                                     │
└─────────────────────────────────────────────────────────────────────────────┘

DEVELOPMENT BRANCH
┌─────────────────────────────────────────────────────────────────────────────┐
│ ●──●──●──●──●──●──●──●──●──●──●──●──●──●──●──●──●──●──●──●──●──●──●──●──● │
│    │        │        │        │        │        │        │        │         │
│    │        │        │        │        │        │        │        │         │
TAGS CREATED AT STABLE POINTS
┌─────────────────────────────────────────────────────────────────────────────┐
│ v0.1.0      v0.2.0      v1.0.0      v1.0.1      v1.1.0      v2.0.0        │
│ (MVP)       (New feat)  (Stable)   (Bug fix)  (New feat)  (Breaking change)│
└─────────────────────────────────────────────────────────────────────────────┘
```

### **Tagging Commands**
```bash
# Create an annotated tag at current commit
git tag -a v1.0.0 -m "Production-ready model with 95% accuracy"

# Tag a specific commit (find SHA with git log --oneline)
git tag -a v1.0.1 a1b2c3d -m "Fix data leakage bug"

# Push tags to GitHub
git push origin --tags

# Delete a tag (local)
git tag -d v0.1.0

# Delete a tag (remote)
git push origin --delete v0.1.0
```

> 📌 **Semantic Versioning Rules**: `MAJOR.MINOR.PATCH`
> - **MAJOR**: Breaking changes (new architecture, incompatible API)
> - **MINOR**: Backward-compatible features (new preprocessing step, additional metrics)
> - **PATCH**: Bug fixes (handling edge cases, documentation updates)

# Git Commands — Quick Reference (Production Use)

## 🔹 Setup & Configuration

```bash
git init
git clone <repo-url>
git config --global user.name "Name"
git config --global user.email "email"
```

---

## 🔹 Daily Workflow (Most Used)

```bash
git status
git add .
git add <file>
git commit -m "message"
git pull
git push
```

---

## 🔹 Viewing History & Changes

```bash
git log --oneline
git log --stat
git diff
git show <commit>
```

---

## 🔹 Branching

```bash
git branch
git branch <name>
git checkout <branch>
git checkout -b <new-branch>
```

---

## 🔹 Merging & Cleanup

```bash
git merge <branch>
git branch -d <branch>
git branch -D <branch>
```

---

## 🔹 Remote Repositories

```bash
git remote -v
git remote add origin <url>
git pull origin main
git push origin main
```

---

## 🔹 Undo & Recovery (Used Often in Production)

```bash
git restore <file>
git restore --staged <file>
git reset --soft HEAD~1
git reset --hard HEAD~1
```

---

## 🔹 Tags (Releases)

```bash
git tag
git tag -a v1.0.0 -m "release"
git push origin v1.0.0
```

---

## 🔹 Inspection & Debugging

```bash
git blame <file>
git reflog
git log --graph --oneline --all
```

---

## 🔹 Common Production Combos

### Update local branch safely

```bash
git pull --rebase
```

### Fix last commit

```bash
git commit --amend
```

### Remove tracked file but keep locally

```bash
git rm --cached file.txt
```
